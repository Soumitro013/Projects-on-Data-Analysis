B. The following data are on speed (sp) of a car and distance (dis) it covers to come to a standstill.
1. Use Glejser test to check for the presence of heteroskedasticity.
2. If heteroskedasticity is present, estimate the parameters assuming the error variance to be a linear function of speed with an intercept term.

```python
import pandas as pd
```

```python
car_data = pd.read_excel('Regression - 2 2.xlsm')
```

```python
car_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb.scatterplot(x='sp',y='dis',data=car_data)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = car_data[['sp']]
Y = car_data[['dis']]
```

```python
lm = LinearRegression()
```

```python
lm.fit(X,Y)
```

```python
PredY = lm.predict(X)
```

```python
lm_Residuals = Y-PredY
```

```python
lm_Residuals
```

```python
plt.scatter(x=lm_Residuals,y=PredY)
plt.annotate('Heteroskedasticity is present',xy = (9,16))
```

```python
Y_Res = abs(lm_Residuals)
```

```python
Y_Res
```

```python
import statsmodels.api as sm
```

1) **|ei| = b0 + b1*X + vi**

```python
X1 = sm.add_constant(X)
```

```python
lm_res1 = sm.OLS(Y_Res,X1).fit()
```

```python
lm_res1.summary()
```

**b̂0** = 3.0097 ,  **b̂1** = 0.3266

2) **|ei| = b0 + b1/X + vi**

```python
X2 = sm.add_constant(1/X)
```

```python
lm_res2 = sm.OLS(Y_Res,X2).fit()
```

```python
lm_res2.summary()
```

**b̂0** = 11.5330 ,  **b̂1** = -30.9150

3) **|ei| = b0 + b1*sqrt(X) + vi**

```python
X3 = sm.add_constant(X**0.5)
```

```python
lm_res3 = sm.OLS(Y_Res,X3).fit()
```

```python
lm_res3.summary()
```

**b̂0** = -1.3810 ,  **b̂1** = 2.5236

See, the **R^2** value is maximum for lm_res1 model.
So, we choose the equation **|ei| = b0 + b1*X + vi** ;
to represent the **Heteroskedasticity**.

We perform **t-test** for **H0** : b̂1 = 0 vs **H1** : b̂1 ≠ 0.
**T** = **b̂1/s1** = 0.3266/0.086 = 3.797674 .
p-val(T) = 0.000352 < 0.05 .
So, the H0 is rejected at 5% level of significance.
i.e. Heteroskedasticity is present.
