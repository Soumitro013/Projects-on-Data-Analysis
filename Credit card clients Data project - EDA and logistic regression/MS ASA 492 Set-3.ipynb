### Creating Dummies for the columns 

```python
#creating dummies for 'SEX' column
sex = pd.get_dummies(data['SEX'])
sex = pd.get_dummies(data['SEX'],drop_first = True) # dropping first column
sex
```

```python
#creating dummies for 'EDUCATION' column
education = pd.get_dummies(data['EDUCATION'])
education = pd.get_dummies(data['EDUCATION'],drop_first = True) # dropping first column
education
```

```python
#creating dummies for 'MARRIAGE' column
marriage = pd.get_dummies(data['MARRIAGE'])
marriage = pd.get_dummies(data['MARRIAGE'],drop_first = True) # dropping first column
marriage
```

```python
#creating dummies for 'PAY_0' column
pay_0 = pd.get_dummies(data['PAY_0'])
pay_0 = pd.get_dummies(data['PAY_0'],drop_first = True) # dropping first column
pay_0
```

```python
#adding all the created dummy variable in to old data frame
data = pd.concat([data,sex,education,marriage,pay_0],axis = 1)
data
```

```python
#dropping irrelavant column
data.drop(['ID','SEX','EDUCATION','MARRIAGE','PAY_0'], axis = 1,inplace = True)
data
```

### Logistic Regression :

```python
x = data.drop("default payment next month",axis = 1)#independent variable
y = data["default payment next month"]#dependent variable
```

#### Splitting data into train and test set

```python
from sklearn.model_selection import train_test_split
#test size 0.25 indicates 75:25 training and test set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 1)
```

#### Fitting Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logi_reg = LogisticRegression()
logi_reg.fit(x_train,y_train)
```

```python
pred=logi_reg.predict(x_test) #making prediction
```

#### Performance checking of the model

```python
logi_reg.score(x_test,y_test) #by using score
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.metrics import confusion_matrix
tn, fp, fn, tp = confusion_matrix(y_test,pred).ravel()
(tn, fp, fn, tp)
```

```python
cnf_matrix = confusion_matrix(y_test,pred)
cnf_matrix
```

```python
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
y_pred_proba = logi_reg.predict_proba(x_test)[::,1]
y_pred_proba
```

```python
#calculate auc 
from sklearn import metrics
auc = metrics.roc_auc_score(y_test, y_pred_proba)
auc
```

```python
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)
auc = metrics.roc_auc_score(y_test, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.legend(loc=4)
plt.show()
```
