## 3. Find the five number summaries

```python
dc = {}
for i in numeric_columns:
    dc[i] = [np.min(possum[i]),np.percentile(possum[i],q=25),np.percentile(possum[i],q=50),np.percentile(possum[i],q=75),np.max(possum[i])]
```

```python
Five_Number_Summaries = pd.DataFrame(dc.values(), index = dc.keys(), columns = ['Minimum','Q1','Median','Q3','Maximum'])
Five_Number_Summaries
```

## 4. Find different correlations

```python
corr_possum = possum[numeric_columns].corr()
corr_possum
```

## 5. Try a correlogram

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
corr_copy = corr_possum.copy()
nc = range(len(numeric_columns))
for i in nc:
    for j in nc:
        if i<j:
            corr_copy.iloc[i,j] = np.nan
        else:
            corr_copy.iloc[i,j] = float('%.2f'%corr_copy.iloc[i,j])
corr_copy
```

```python
fig = sns.heatmap(corr_copy,cmap='coolwarm',annot = True)
b,t = fig.get_ylim()
fig.set_ylim(b+0.5,t-0.5)
plt.title('Correlogram')
plt.show()
```

## 9. Try different interactive charts 

```python
possum_3 = possum.groupby('site').mean()
possum_3
```

```python
fig,ax = plt.subplots(3,3,figsize = (15,15),gridspec_kw = {'hspace':0.5})
fig.suptitle('\nBar Diagrams',fontsize = 20)
sns.barplot(ax=ax[0][0],data=possum_3,x=possum_3.index,y='skullw',palette = 'pastel')
ax[0][0].set_title('Skull Weight')
sns.barplot(ax=ax[0][1],data=possum_3,x=possum_3.index,y='hdlngth',palette = 'pastel')
ax[0][1].set_title('Head Length')
sns.barplot(ax=ax[0][2],data=possum_3,x=possum_3.index,y='footlgth',palette = 'pastel')
ax[0][2].set_title('Foot Length')
sns.barplot(ax=ax[1][0],data=possum_3,x=possum_3.index,y='totlngth',palette = 'pastel')
ax[1][0].set_title('Total Length')
sns.barplot(ax=ax[1][1],data=possum_3,x=possum_3.index,y='taillgth',palette = 'pastel')
ax[1][1].set_title('Tail Length')
sns.barplot(ax=ax[1][2],data=possum_3,x=possum_3.index,y='earconch',palette = 'pastel')
ax[1][2].set_title('Earconch')
sns.barplot(ax=ax[2][0],data=possum_3,x=possum_3.index,y='eye',palette = 'pastel')
ax[2][0].set_title('Eye')
sns.barplot(ax=ax[2][1],data=possum_3,x=possum_3.index,y='belly',palette = 'pastel')
ax[2][1].set_title('Belly')
sns.barplot(ax=ax[2][2],data=possum_3,x=possum_3.index,y='chest',palette = 'pastel')
ax[2][2].set_title('Chest')
plt.show()
```
